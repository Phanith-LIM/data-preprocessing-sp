```python
import pandas as pd
import pandasql as ps # type: ignore
from natsort import natsorted # type: ignore
import os
```

```python
path = '/Users/PhanithLIM/Documents/05.Dataset/Speech Recognition/processing/WMC-Internation/clean-logs'
files = os.listdir(path)
files = [f for f in files if f.endswith('.csv')]
files = natsorted(files)
files = [os.path.join(path, f) for f in files]
files
```

```python
list_df = [ pd.read_csv(f) for f in files ]
df = pd.concat(list_df, axis=0)
df = df.reset_index(drop=True)
df = df.drop(columns=['id', 'link'])
df.head(10)
```

```python
q1 = '''
SELECT 
    *,
    CASE 
        WHEN speaker LIKE 'កញ្ញា%' THEN 'ស្រី'  
        ELSE 'ប្រុស'
    END AS gender
FROM df; 
'''
df = ps.sqldf(q1, locals())
df.head(10)
```

```python
q2 = '''
SELECT COUNT(*) AS count_audio, speaker, 
    CASE 
        WHEN speaker LIKE 'កញ្ញា%' THEN 'ស្រី'
        ELSE 'ប្រុស'
    END AS gender 
FROM df
GROUP BY speaker
'''
result = ps.sqldf(q2, locals())
result
```

```python
q3 = f'''
SELECT COUNT(*) AS count, gender
FROM ({q2}) AS derived_table
GROUP BY gender;
'''
result = ps.sqldf(q3, locals())
result
```
